## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
city_data=pd.DataFrame({"City":name,
                       "Cloudiness":cloudiness,
                       "Country":country,
                       "Date": date,
                       "Humidity": humidity,
                       "Lat": lat,
                       "Lng": lng,
                       "Max Temp": max_temp,
                       "Wind Speed": wind_speed})

city_data.head()
```

#### Latitude vs. Temperature Plot

```python
today=time.strftime("%m/%d/%Y")
city_data_filtered=city_data.loc[city_data["Humidity"]<=100]
plt.scatter(city_data_filtered["Lat"], city_data_filtered["Max Temp"], color='steelblue',edgecolor='black',s=40)

plt.xlabel("Latitude")
plt.ylabel("Max Temperature (F)")
plt.title("City Latitude vs. Max Temperature ("+today+")")
plt.grid()
plt.show()
```

```python
city_data_filtered.plot(kind="scatter", x="Lat",y="Humidity",color="steelblue",edgecolor="black",s=40)
plt.xlabel("Latitude")
plt.ylabel("Humidity (%)")
plt.title("City Latitude vs. Humidity ("+today+")")
plt.grid()
plt.show()
```

```python
city_data_filtered.plot(kind="scatter", x="Lat",y="Cloudiness",s=40,color="steelblue", edgecolor="black")
plt.xlabel("Latitude")
plt.ylabel("Cloudiness (%)")
plt.title("City Latitude vs. Cloudiness ("+today+")")
plt.grid()
plt.show()
```
